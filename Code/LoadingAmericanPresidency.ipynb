```python
import json
import pandas as pd

file_path = r"C:\Users\shime\Downloads\presidency.json"

# Load JSON
with open(file_path, 'r', encoding='utf-8') as f:
    data = json.load(f)

rows = []

# Loop over presidents
for president, categories in data.items():
    # Each president has multiple types (Written, Debates, etc.)
    for category, speeches in categories.items():
        if isinstance(speeches, list):  # Only proceed if it’s a list of speeches
            for speech in speeches:
                rows.append({
                    "President": president,
                    "Category": category,
                    "Title": speech.get("title") or speech.get("document", ""),
                    "Date": speech.get("date"),
                    "Text": (
                        speech.get("text")
                        or speech.get("speech")
                        or speech.get("content")
                        or ""
                    )
                })

# Convert to DataFrame
df_presidency = pd.DataFrame(rows)
print(df_presidency.shape)
print(df_presidency.head())
```

```python
import pandas as pd
import json

path = r"C:\Users\shime\Downloads\presidency.json"

# Load the JSON file (assuming it's in your current working directory)
with open(path, "r") as f:
    data = json.load(f)

rows = []
for president, categories in data.items():
    for category, docs in categories.items():
        if isinstance(docs, list) and len(docs) > 0:
            for doc in docs:
                rows.append({
                    "President": president,
                    "Category": category,
                    "DocCategory": doc.get("category"),
                    "Subcategory": doc.get("subcategory"),
                    "Title": doc.get("title"),
                    "Date": doc.get("document_date"),
                    "Text": doc.get("content")
                })

# Create DataFrame
df = pd.DataFrame(rows)
print(df.shape)
print(df.head(3))
```
